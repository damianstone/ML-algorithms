### Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('position_salaries.csv')

# features (from the second column since the first is the same as the second but as categorical)
x = dataset.iloc[:, 1:-1].values

# dependent variable (last column)
y = dataset.iloc[:, -1].values
```

### Training the linear regression

```python
from sklearn.linear_model import LinearRegression

regression_model_1 = LinearRegression()

regression_model_1.fit(x, y)
```

### Training the polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures

# to add an exponential to x1
# degree = 2 ->  get a new feature matrix containing [1, x, x^2], where 1 is the bias term, x is the original feature, and x^2 is the new polynomial feature
polynomial_regression = PolynomialFeatures(degree=2)
x_poly = polynomial_regression.fit_transform(x)

# new linear regression using the x_poly
regression_model_2 = LinearRegression()
regression_model_2.fit(x_poly, y)
```

### Plot linear regression

```python
plt.scatter(x, y, color="red")
plt.plot(x, regression_model_1.predict(x), color="blue")
plt.title("Linear regression")
plt.xlabel("Pisition level")
plt.ylabel("Salary")
plt.show()
```

### Plot polynomial regression

```python
plt.scatter(x, y, color="red")
plt.plot(x, regression_model_2.predict(x_poly), color="blue")
plt.title("Plynomial regression")
plt.xlabel("Pisition level")
plt.ylabel("Salary")
plt.show()
```

### Predicting single values

```python
# Predicting a new result with Linear Regression
regression_model_1.predict([[6.5]])

# Predicting a new result with Polynomial Regression
regression_model_2.predict(polynomial_regression.fit_transform([[6.5]]))
```
